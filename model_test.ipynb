```python
import torch
import cv2

video_path = 'C:/Users/osman/OneDrive/Masaüstü/Model_Test/2021.mp4'
video_path_out = 'C:/Users/osman/OneDrive/Masaüstü/Model_Test/output_video1.mp4'

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'XVID'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

model = torch.hub.load('ultralytics/yolov5:master', 'custom', path=r'C:/Users/osman/OneDrive/Masaüstü/Model_Test/model_ağırlıkları/250_epoch_transformer/best.pt')

threshold = 0.5

while ret:
    results = model(frame)

    for result in results.xyxy[0].cpu().numpy():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, model.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
from ultralytics import YOLO
video_path = 'C:/Users/osman/OneDrive/Masaüstü/Model_Test/2021.mp4'
video_path_out = 'C:/Users/osman/OneDrive/Masaüstü/Model_Test/output_video1.mp4'


cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H,W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))


model = YOLO('C:/Users/osman/OneDrive/Masaüstü/Model_Test/yolov8/200 epoch/best (2).pt') 

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
